```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

from mlxtend.plotting import plot_decision_regions

lr = LogisticRegression(C=5.0, random_state=42, n_jobs=-1) # add some params
svm = SVC(kernel='linear',C=5.0, random_state=42) # here too

### YOUR CODE HERE
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(lr, X, y, cv=5, scoring='accuracy')
```

```python
np.mean(scores)
```

```python
cv = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
```

```python
def plot_decisions(X, y, clf):
    
    name = type(clf).__name__
    
    acc_score = cross_val_score(clf, X, y, cv=cv, scoring='accuracy')
    print('{:30}: accuracy = {:.4f}'.format(name, np.mean(acc_score)))
    
    clf.fit(X, y)
    plt.figure(figsize=(7, 5))
    plt.title(name, fontsize=20)
    plot_decision_regions(X, y, clf, legend=1)
```

```python
models = [lr, svm]
```

**_<font color=blue>Conclusions:</font>_**

- The classification quality improve with all kernels, except for kernel='sigmoid'.
- Decision boundaries:
    - For `kernel='poly'`. The decision boundary is what we'd expect when trying to find a decision boundary as a polynomial of degree $> 2$ (that is, by adding polynomial features (with polynomial kernel we avoid to add them)).
    - For `kernel='rbf'`. Here we got a small value of $\gamma$, which explains the smoothness of the decision boundary. Moreover, because of the $\gamma$ and $C$ values, the observations have a larger range of influence with a smaller overal margin. That why the objects of class 1 end up being enclosed in subregion of the plane (for other $C$ values that subregion will be greater).
    - For `kernel='sigmoid'`, the decision boundary is better than linear but still not better that using rbf and polynomial kernel.
- For the chosen degree see below.

Not let's play with the degree hyperparameter for the polynomial kernel:

```python
from sklearn.preprocessing import PolynomialFeatures

### YOUR CODE HERE
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline(
    [
        ('poly_features', PolynomialFeatures(include_bias=False)),
        ('log_reg', LogisticRegression(C=5.0, random_state=42, n_jobs=-1))
    ]
)
```

```python
params_pipe = {
    'poly_features__degree': np.arange(2, 6)
}
```

```python
grid_search = GridSearchCV(
    pipeline,
    params_pipe,
    cv=5,
    scoring='accuracy'
)

grid_search.fit(X, y)
```

```python
grid_search.best_params_
```

```python
svm_poly = SVC(kernel='poly', degree=4, coef0=2.0, gamma='scale')

for model in [grid_search.best_estimator_, svm_poly]:
    plot_decisions(X, y, model)
```

```python
print(grid_search.best_estimator_[0].get_feature_names(['x', 'y']))
```

**_<font color=blue>Conclusions:</font>_**

- The decision boundary is quite similar to the result we got using a SVM with polynomial kernel and hence the comparable accuracy scores.
- We use a $4$-th degree polynomial. It's the best hyperparameter value found with grid search.
- On the border type for the polynomial logistic regression model: 
    - $L = \sum\limits_i^n \log\sigma(y_ix_i^{T}w)$ and the decision boundary is given by $x_i^{T}w = c$ where $c$ is the threshold we choose for classification (usually $c = 0.5$).
    - For a simple logistic regression model $x_i = (x_1^i, x_2^i)$ and the decision boundary is given by the line $x_1^iw_1 + x_2^iw_2 = c$.
    - After `PolynomialFeatures` we can see clearly that the decision boundary resemble a $4$-th degree polynomial (the best degree found) with a new feature vector $\tilde{x_i}$ with components given by:

```python
print(grid_search.best_estimator_[0].get_feature_names(['x1', 'x2']))
```

- The difference in decision boundary shapes may be given by the way `PolynomialFeatures` and polynomial kernel generate a, in this case, $4$-th degre polynomial:
    - `PolynomialFeatures`: by computing all combinations **of features** up to $4$ degree.
    - Polynomial kernel: by raising to the $4$-th **the similarity measure $\langle \cdot,\cdot \rangle$** between observations pairs: $(\gamma \langle x_i, x_j\rangle + r)^4 = (\gamma x_i^{\top}x_j + r)^4$

**_<font color=blue>Observations:</font>_**

- Since the data is simple and visually easy to interpret, aside from Logistic Regression and SVM (with rbf kernel) we'll train a kNN model.

```python
X_2, y_2 = points, labels
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr_pipe = Pipeline(
    [
        ('poly_features', PolynomialFeatures(include_bias=False)),
        ('log_reg', LogisticRegression(random_state=42, n_jobs=-1))
    ]
)

svm_rbf = SVC(kernel='rbf', random_state=42)

rf = RandomForestClassifier(max_depth=8, random_state=42, warm_start=True, n_jobs=-1)

knn = KNeighborsClassifier()
```

```python
params_lr = {
    'poly_features__degree': np.arange(2, 6),
    'log_reg__C': [0.1, 1.0, 5.0, 7.0],
}

params_svm_rbf = {
    'C': [1e-3, 1.0, 5.0],
    'gamma': ['scale', 'auto'],
}

params_rf = {
    'n_estimators': np.arange(6, 21),
}

params_knn = {
    'n_neighbors': np.arange(5, 15),
    'weights': ['uniform', 'distance'],
    'metric': ['braycurtis', 'minkowski']
}
```

```python
models = [lr_pipe, svm_rbf, rf, knn]
params_list = [params_lr, params_svm_rbf, params_rf, params_knn]
```

```python
from itertools import product
```

```python
list(product([0,1], repeat=2))
```

```python
gs = gridspec.GridSpec(2, 2)
fig = plt.figure(figsize=(20,15))

best_models = []

for i, model, params, grd in zip(np.arange(0,4), models, params_list, product([0, 1], repeat=2)):
    params = params_list[i]
    grid_search = GridSearchCV(model, params, cv=cv, scoring='accuracy', n_jobs=-1, refit=True)
    grid_search.fit(X_2, y_2)

    name = type(model).__name__
    
    if name == 'Pipeline':
        name = 'LogisticRegression ' + name
    
    print('\n{}'.format(name))

    print('\tBest Parameters:')
    for param, value in grid_search.best_params_.items():
        print("\t\t{:20}: {}".format(param, value))
    print('\tAccuracy score: {:.4f}'.format(grid_search.best_score_))

    best_models.append(grid_search.best_estimator_)

    ax = plt.subplot(gs[grd[0], grd[1]])

    plt.title('{}'.format(name), fontsize=20)
    plot_decision_regions(X_2, y_2, clf=grid_search.best_estimator_, legend=2)
```

**_<font color=blue>Observations:</font>_**

- The best models happen to be Polynomial Logistic Regresion and Poly SVM. As we saw above, these model are more flexible when dealing with the geometric structure of the data.
- kNN and Random Forest do comparatively worse because aren't able to generalize well due to the data geometry:
    - _Random forest:_ to deal with missclassification we could increase the trees' depth, but at the current number we already notice signs of overfitting, specially in the region between green and red classes.
    - _kNN:_ the prediction method used is local. That's why we get non-smooth boundaries. On the other hand, in some regions observations of different classes are sort of jammed together, which makes difficult for kNN to predict the class of an observation by hard-voting (that's why it ends up with a high number of neighbors).
